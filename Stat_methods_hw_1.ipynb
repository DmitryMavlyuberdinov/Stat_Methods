## Вопрос №2

По всей видимости речь идет об подозрении в убийстве жены. То есть событие $A$ - бил, $B$ - убил. Прокурор говорит об условной вероятности события $P(B|A)$, то есть какова вероятность того, что O. J. Simpson убил при условии, что ранее он бил. Адвокат переводит все в рассмотрение другой условной вероятности $P(A|B)$, что не одно и тоже с $P(B|A)$. Взаимосвязь этих вероятностей опеределяется по формуле Байеса:
$$
P(B|A) = \frac{P(A|B)P(B)}{P(A)}
$$
То есть судья ошибся принимаю во внимание доводы адвоката.

Рассуждая теоретически: абсолютная погрешность убывает со скоростью $\sqrt(n)$. 

Результат отклонения среднего измеренного значения от истинного значения веса:
$<x> - \sigma_{<x>} \leq \bar x \leq  <x> + \sigma_{<x>} $
Отклонение среднего измеренного значения от истинного значения веса, по условию задачи не должно превышать массу атома железа или $ 9,3 \cdot 10^{-26} $

Так как нахождение значения между двумя рисками циферблата равномерно распределено, то дисперсия конкретного измерения $$\sigma_{x} = \frac{a - b}{2\sqrt(3)} = 0,29$$
$$ n = \left( \frac{\sigma_{x}}{\sigma_{<x>}}\right) ^2 = \left( \frac{0,29}{9,3 \cdot 10^{-26}} \right) ^{2} = 3 \cdot 10^{50}  $$

## Задача №3

Частотный подход:
Обозначим вероятность выигрыша Алисы, как $ p $. Тогда вероятность выигрыша Боба, это три проигрыша Алисы:
    $$ P(Bob) = (1-p)^3$$
Вероятность выигрыша Алисы тогда, есть результат не выигрыша Боба:
    $$ P(Alisa) = 1 - P(Bob) = 1 - (1-p)^3$$
Вероятность $p$ можно оценить данных эмпирических данных, как $p= \frac{5}{8}$.
Тогда получаем:
    $$ P(Alisa) =  1 - (1-p)^3 = 1 - \left(1 - \frac{5}{8}\right)^3 = 1 - \frac{27}{512} =  \frac{485}{512} = 0,947 $$
Байесовский подход:
На сколько наша оценка вероятности $p$ верна? Рассмотрим условную вероятность, оценки вероятности $p$ победы Алисы после того, как событие из условия задачи произошло:
    $$ P(p|A = 5, B = 3) $$
Тогда, ввиду непрерывности изменения вероятности в диапазона $ p = 0 \dots 1 $, получаем математическое ожидание вероятности победы Боба:
    $$ E(Bob) = \int_0^1 (1-p)^3 P(p|A = 5, B = 3) \,\mathrm{d}p $$
По формуле Байеса, определим вероятность $P(p|A = 5, B = 3)$:
$$P(p|A = 5, B = 3) = \frac{P(A = 5, B = 3|p)P(p)}{\int_0^1 P(A = 5, B = 3|p)P(p) \,\mathrm{d}p}$$ (1)
Предположим, что плотность распределения вероятности $p(p)$ распределена по равновероятному закону, а значит 
$P(p) =\int_0^p \frac{1}{1-0} \,\mathrm{d}p = p $.
$$P(A = 5, B = 3|p)P(p) =  \binom{8}{5} p^5 q^3 = \frac{8!}{5!3!}p^5 (1-p)^3 $$
 $$ E(Bob) = \frac{\int_0^1 (1-p)^6 p^6 \,\mathrm{d}p}{\int_0^1 (1-p)^3 p^6 \,\mathrm{d}p} $$
Воспользуемся формулой:
$$ \int_0^1 (1-p)^{m-1} p^{n-1} \,\mathrm{d}p = \frac {\Gamma(m)\Gamma(n)}{\Gamma(m+n)} $$
$$ \Gamma(n+1) = n!$$
Тогда получаем:
$$E(Bob) = \frac{ \frac{6!6!}{13!}}{\frac{6!3!}{10!}} = \frac{10}{143} $$
И вероятность победы Алисы:
$$P(Alisa) = 1 - \frac{10}{143} = 0,93 $$
Видим, что вероятность победы Алисы чуть ниже, чем в случае апостериорно вычисленной вероятности. Заметим, что в случае если $p=0,5$, мы получим $P(Alisa) =  1 - (1-p)^3 = 1 - \frac{1}{8}  = \frac{7}{8} = 0.875$. То есть байесовский подход более аккуратно учитывает результат случайно возникающей апостериорной вероятности.


## Задача №4

Обозначим:

$A$ - человек болен, $P(A) = \frac{1}{140}$;

$(-|A)$ - человек получил отрицательный тест, при том, что он болен, $P(-|A)=0,08$;

$(-|\bar A)$ - человек получил отрицательный тест, при том, что он не болен, $P(-|\bar A)=1 - P(-|A)=0,92$;

$(+|\bar A)$ - человек получил положительный тест, при том, что он не болен, $P(+|\bar A)=0,02$;

$(+|A)$ - человек получил положительный тест, при том, что он болен, $P(+|A)=1-P(+|\bar A) = 0,98$;


Необходимо найти вероятность события: $P(A|++-)$. 
По формуле Байеса:
$$
P(A|++-) = \frac{P(++-|A)P(A)}{P(++-)}
$$
События $P(+|A)$ и $P(-|A)$ - независимые, тогда по теореме умножения: 
$$
P(++-|A) = P(-++|A) = P(+|A)P(+|A)P(-|A) = 0,98 \cdot 0,98 \cdot 0,08
$$
Для определения вероятности $P(++-)$, разобьём пространство исходов на случае, когда человек болен и не болен.
Тогда получим:
$$
P(++-) = P(++-|A)P(A) + P(++-|\bar A)P(\bar A) = P(+|A)P(+|A)P(-|A)P(A) + P(+|\bar A)P(+|\bar A)P(-|\bar A)P(A)P(\bar A)=
0,98 \cdot 0,98 \cdot 0,08 \cdot \frac{1}{140} + 0,02 \cdot 0,02 \cdot 0,92 \cdot \frac{139}{140} = 0.0005488 + 0.0003653 = 0.000914
$$
$$
P(A|++-) = \frac{0.0005488}{0.000914} = 0.60
$$
