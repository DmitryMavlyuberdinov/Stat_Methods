##  Вопрос №1

Пусть задано вероятностное пространство $(\Omega ,{\mathcal {F}},\mathbb {P} )$, и на нём определена случайная величина $ ξ:\Omega \to \mathbb {R}$. Тогда:
$\forall x\in {\mathbb {R}} : \{\omega \in \Omega : ξ(\omega) \leq x\} \in {\mathcal {F}}$

Тогда вероятность попадания в отрезок до $x$ включительно:

$${P}(ξ \leq x) = {P}(\{\omega \in \Omega : ξ(\omega) \leq x\}) = F_{ξ}(x)$$

Функция $F_{ξ}(x)$ называют функция распределения случайной величины, часто обозначают, как CDF. 
Множество отрезков от минус бесконечности до $x$ - называют Борелевым множеством и обозначают ${\mathcal {B}}$.


Функция распределения $F_{ξ}(x)$ любой случайной величины удовлетворяет следующим трем свойствам:

1) $F_{ξ}$ — функция неубывающая;
2) $\lim _{x\to -\infty }F_{ξ}(x)=0,\;\lim _{x\to \infty }F_{ξ}(x)=1$;
3) $F_{ξ}$ непрерывна справа.

Бывают два типа распределение случайной величины дискретное и непрерывное.

```python
from scipy.integrate import simps
def integrateS(func, a, b, numPoints = 1000):
    x = np.linspace(a,b, numPoints, endpoint=True)
    return simps(x = x, y = func(x))
```

```python
convolute = lambda f, g, a, b:  np.vectorize(lambda x:integrateS(lambda y: f(x-y)*g(y), a, b))
```

```python
wStep = 4 # Ширина ступеньки
sigma = 0.5 # Стандартное отклонение Гауссианы
sigma_f = 0.8
mu_cauchy = 0 
gamma = 1

# ступенька
#f = lambda x: 1/wStep if (-wStep / 2 <= x <= wStep / 2) else 0.0 # нормировка на единицу
#f = lambda x : 1/np.sqrt(2*math.pi)/sigma_f*np.exp(-x*x/2/sigma_f/sigma_f)
f = lambda x : 1/math.pi * gamma/ ((x-mu_cauchy)*(x-mu_cauchy)+gamma*gamma)
f = np.vectorize(f)

# Функция Гаусса
g = lambda x : 1/np.sqrt(2*math.pi)/sigma*np.exp(-x*x/2/sigma/sigma)

# Plot plot = new Plot(title: "Convolution test plot", showLegend: true)


# автоматическое определение окна интегрирования и отрисовки
wind = max(2*wStep, 4*sigma)
x = np.linspace(-wind, wind, 200)

traces = [
    go.Scatter(x = x, y = f(x), name = "f(x)"),
    go.Scatter(x = x, y = g(x), name = "g(x)"),
    go.Scatter(x = x, y = convolute(f,g, -wind, wind)(x), name = "convolution")
]

go.Figure(traces)
```
